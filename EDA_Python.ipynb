```python
def clasificar_fallo(real, pred):
    if real == 1 and pred == 1:
        return 'True Positive'
    elif real == 0 and pred == 0:
        return 'True Negative'
    elif real == 0 and pred == 1:
        return 'False Positive'
    elif real == 1 and pred == 0:
        return 'False Negative'

df_clave['compas_outcome'] = df_clave.apply(lambda row: clasificar_fallo(row['two_year_recid'], row['compas_pred']), axis=1)
```

🔍 ¿Qué muestra el DecileScore de COMPAS?

El DecileScore es el puntaje de riesgo que asigna COMPAS a una persona evaluada, y va del 1 al 10, donde:

    1 = riesgo muy bajo de reincidencia

    10 = riesgo muy alto de reincidencia

Este score se calcula en base a un cuestionario automatizado, considerando:

    Edad

    Número de delitos previos

    Historial de faltas

    No necesariamente el delito actual, ni la raza (aunque el resultado puede estar correlacionado con la raza).
